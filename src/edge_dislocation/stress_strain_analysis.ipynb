```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Extract the data from the text file
fileName = "stress_strain.txt"

stressStrainData = pd.read_csv(fileName, delim_whitespace=True, header=None, skiprows=1) # Read the data from the textfile and skip the first row

stressStrainData.columns = ['Strain', 'Stress_X', 'Stress_Y', 'Stress_Z'] # Append the column headers

stressStrainData.head()
```

```python
# Plot the data on a graph
plt.figure(figsize=(8,6))
plt.plot(stressStrainData['Strain'], stressStrainData['Stress_X'], label='Stress Strain Curve X axis', color='blue', linestyle='--')
plt.plot(stressStrainData['Strain'], stressStrainData['Stress_Y'], label='Stress Strain Curve Y axis', color='red')
plt.plot(stressStrainData['Strain'], stressStrainData['Stress_Z'], label='Stress Strain Curve Z axis', color='green')

plt.title('Stress Strain Curve for LAMMPS Simulation at 300K')
plt.xlabel('Strain')
plt.ylabel('Stress')
plt.legend(fontsize=8)

plt.savefig('Stress Strain Curve', dpi=500)

plt.show()

```
